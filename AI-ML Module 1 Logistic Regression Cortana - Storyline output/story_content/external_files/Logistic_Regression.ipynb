```python
import numpy as np
import pandas as pd
```

```python
dataframe = pd.read_csv('ionosphere.csv')
```

```python
dataframe
```

```python
dataframe['Actual_Class'] = np.where(dataframe['Class'] == 'good', 0, 1)
```

```python
dataframe
```

```python
dataframe = dataframe.drop(columns = ['Class'])
```

```python
dataframe
```

```python
y = dataframe['Actual_Class']
X = dataframe.drop(columns = ['Actual_Class'])
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)
```

```python
X_test_pred = classifier.predict(X_test)
```

```python
X_test_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy = np.around(accuracy_score(X_test_pred, y_test)*100, 2)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, X_test_pred)
disp = ConfusionMatrixDisplay(confusion_matrix = cm)
disp.plot()
```
